```python
# Import library googleapiclient.discovery untuk menggunakan YouTube Data API
import googleapiclient.discovery
# Import library pandas sebagai pd untuk manipulasi data menggunakan DataFrame
import pandas as pd

# Konfigurasi nama layanan API dan versi yang akan digunakan
api_service_name = "youtube"
api_version = "v3"
# Masukkan kunci pengembang Anda
DEVELOPER_KEY = dev

# Membangun objek YouTube API dengan menggunakan kunci pengembang dan versi API yang ditentukan
youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)

# Membuat permintaan (request) untuk mendapatkan data komentar dari video tertentu
request = youtube.commentThreads().list(
    part="snippet",
    videoId=video_id,  # ID video yang akan diambil komentarnya
    maxResults=1999  # Jumlah maksimum komentar yang akan diambil
)

# Inisialisasi list kosong untuk menyimpan komentar
comments = []

# Mengirimkan permintaan dan mendapatkan respons
response = request.execute()

# Mendapatkan komentar dari respons
for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']
    public = item['snippet']['isPublic']
    # Menambahkan komentar ke dalam list comments dalam bentuk list
    comments.append([
        comment['authorDisplayName'],  # Nama penulis komentar
        comment['publishedAt'],  # Waktu publikasi komentar
        comment['likeCount'],  # Jumlah suka komentar
        comment['textOriginal'],  # Isi komentar
        public  # Status publikasi komentar
    ])

# Melakukan iterasi untuk mengambil komentar dari halaman berikutnya jika ada
while (1 == 1):
    try:
        nextPageToken = response['nextPageToken']  # Mengambil token untuk halaman berikutnya
    except KeyError:
        break  # Keluar dari loop jika tidak ada halaman berikutnya
    
    nextPageToken = response['nextPageToken']  # Mengambil token untuk halaman berikutnya
    
    # Membuat permintaan baru dengan menggunakan token halaman berikutnya
    nextRequest = youtube.commentThreads().list(
        part="snippet", 
        videoId=video_id,  # ID video (opsional jika tetap sama dengan sebelumnya)
        maxResults=1999,  # Jumlah maksimum komentar per halaman
        pageToken=nextPageToken  # Token untuk halaman berikutnya
    )
    # Mengirimkan permintaan baru dan mendapatkan respons
    response = nextRequest.execute()
    # Mendapatkan komentar dari respons dan menambahkannya ke dalam list comments
    for item in response['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        public = item['snippet']['isPublic']
        comments.append([
            comment['authorDisplayName'],
            comment['publishedAt'],
            comment['likeCount'],
            comment['textOriginal'],
            public
        ])

# Membuat DataFrame dari list comments dengan kolom-kolom yang ditentukan
df = pd.DataFrame(comments, columns=['author', 'updated_at', 'like_count', 'text', 'public'])
# Menampilkan informasi DataFrame seperti jumlah baris dan tipe data kolom
df.info()
```

```python
import json

# Membuat list of dictionaries dengan key yang diinginkan
comments_data = []
for comment in comments:
    comment_dict = {
        'video': video_id,
        'author': comment[0],
        'updated_at': comment[1],
        'like_count': comment[2],
        'text': comment[3],
        'public': comment[4]
    }
    comments_data.append(comment_dict)

# Membuat dictionary yang berisi key 'comments' dengan nilai dari comments_data
data = {'comments': comments_data}

# Menyimpan dictionary ke dalam file JSON
with open('comments.json', 'w') as file:
    json.dump(data, file, indent=4)
```

## Menyimpan json ke dalam MongoDB

```python
import json
from pymongo import MongoClient

# Koneksi ke MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["youtube_rewind_indonesia"]
collection = db["collect_2016"]

# Membaca file JSON
with open("comments.json", "r") as file:
    data = json.load(file)
    comments = data["comments"]

# Menyisipkan data ke MongoDB
collection.insert_many(comments)

print("Data berhasil diimpor ke MongoDB.")
```
