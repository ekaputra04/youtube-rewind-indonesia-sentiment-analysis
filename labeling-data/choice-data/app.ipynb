```python
import pandas as pd
import os
from pymongo import MongoClient
from dotenv import load_dotenv
```

```python
# Nama database dan collection
db_name = 'youtube_rewind_indonesia'
collection_name = 'full_labeling_2015'

# Memuat value dari file .env
load_dotenv()

mongodb_url = os.getenv('URL_SANDY')
local_url = os.getenv('URL_LOCAL')
```

```python
# Membuat koneksi ke MongoDB
client = MongoClient(local_url)
db = client[db_name] 
full_labeling_collection = db[collection_name]
```

```python
labeled_data = list(full_labeling_collection.find())
df_labeled = pd.DataFrame(labeled_data)
```

```python
# Menghitung jumlah setiap label
label_counts = df_labeled['predicted_label'].value_counts()
label_counts
```

```python
# Menyimpan jumlah ke dalam variabel
positif_count = label_counts.get('positif', 0)
negatif_count = label_counts.get('negatif', 0)
netral_count = label_counts.get('netral', 0)

# Menampilkan hasil
print(f"Jumlah komentar positif: {positif_count}")
print(f"Jumlah komentar negatif: {negatif_count}")
print(f"Jumlah komentar netral: {netral_count}")
```

```python
# Membagi dataframe menjadi tiga berdasarkan nilai dari kolom 'predicted_label'
df_positif = df_labeled[df_labeled['predicted_label'] == 'positif']
df_negatif = df_labeled[df_labeled['predicted_label'] == 'negatif']
df_netral = df_labeled[df_labeled['predicted_label'] == 'netral']
```
